## Lists

````
a = [0, 4, 2, 8, 9]  # Create a list with initial items
b = []  # Create an empty list

# Length
print(len(a))  # 5

# The same indexing operations on strings work for lists
print(a[1])    # 4
print(a[2:4])  # [2, 8]
print(a[-2])   # 8

# Sort the list
print(sorted(a))                # [0, 2, 4, 8, 9] Creates a copy, does not modify the original list
print(sorted(a, reverse=True))  # [9, 8, 4, 2, 0] Also a copy

# Add items to the end of the list
a.append(10)  # Returns nothing
print(a)      # [0, 4, 2, 8, 9, 10]
a += [1, 12]  # Returns nothing
print(a)      # [0, 4, 2, 8, 9, 10, 1, 12]

# Lists don't have to be the same type in python
a.append('machine learning')
print(a)            # [0, 4, 2, 8, 9, 10, 1, 12, 'machine learning']
# print(sorted(a))  # Causes an error

# Searching
print(8 in a)    # True
print(100 in a)  # False
print(a.index(8))
````

## List Comprehension
A cool and extremely useful feature of Python! Good for condensing code using loops and conditionals.

The general syntax is:

```
[ <expression> for <name> in <list> if <filter> ]
```

List comprehension also applies to dictionaries and sets, as follows.

```
{ <expression> for <key>, <value> in <dict> if <filter> }
```

````
# Sometimes code can be simplified using list/dictionary comprehensions
def add_one(x):
  return x + 1

# The standard way to apply a function to every element and save in y
x = [0, 1, 2, 3]
y = []
for i in range(len(x)):
  y.append(add_one(x[i]))

# This is equivalent to the above
y = [add_one(x_i) for x_i in x]
# funtion to element for element in data structure

# Add conditionals
y2 = [add_one(x_i) for x_i in x if x_i > 1]
# function to element for element in data structure if element meets conditional

# Dictionary comprehensions can be used to create dictionaries easily.
# This creates a mapping from x_i -> x_i * 4
d = {x_i : x_i * 4 for x_i in x}
# element key : function to element value for element in data structure
print(d)
print(d[2])  # 8
````

## Function with Arguments
### Default Values for Arguments


*   You can provide default values for a function's arguments
*   These arguments are optional when the function is called
*   Non-default argument should always precede default arguments; otherwise, it reportsSyntaxError



````
def myfun(a, b=3, c='hello'):
  return a+b

# To call the function, I can pass in different arguments
myfun(5, 3, 'bob') # should print 8

myfun(5, 3) # should print 8

myfun(5) # should print 8
````

### Keyword Arguments
*   Functions can be called with arguments out of order
*   These arguments are specified in the call
*   Keyword arguments are called with variable = value format
*   Keyword arguments can be used after all other arguments

````
def myfun(a, b, c):
  return a - b
````

````
myfun(2, 1, 43) # should return 1
````

````
# specified in any order
myfun(c=43, b=1, a=2) # should return 1
````

````
# Keyword arguments can be used after all other arguments
myfun(2, c=43, b=1)
````

````
# should report error
myfun(a=2, b=3, 5)
````

### *args
*   Suppose you want to accept a variable number of non-keyword arguments to your function, you can use *args

````
def print_things(*args):
  """ args is a tuple fo arguments passed to the function"""
  print(args)
````

````
print_things('a','b','c')
````

````
lst = ['a','b','c']
print_things(*lst)
````

### **kwargs
*  Suppose you want to accept a variable number of keyword arguments to your function, you can use **kwargs

````
def print_keywords(**kwargs):
  print(kwargs)
````

````
print_keywords(first_name="Tom", last_name = 'Cruise')
````

````
dict1 = {'first_name': 'Tom', 'last_name': 'Cruise'}
print_keywords(**dict1)
````

### All together: positional, *args and **kwargs

````
def myfun(positional, *args, **kwargs):
  print(positional)
  if args:
    print(args)
  if kwargs:
    print(kwargs)
````

````
myfun('hi')
````

````
myfun('hi', 1,2,3)
````

````
myfun('hi', 1,2,3, a='hello', b='class')
````

## Apply

Apply is a very powerful method which can be used for making major data manipulation tasks. Much faster than standard for loops because of internal optimizations.

NOTE: In some assignments, your code could never finish running if you use for loops due to the size of the datasets!

````
%%timeit

countries = pd.DataFrame({'country': ['Belgium', 'France', 'Germany', 'Netherlands', 'United Kingdom'],
        'population': [11.3, 64.3, 81.3, 16.9, 64.9],
        'area': [30510, 671308, 357050, 41526, 244820],
        'capital': ['Brussels', 'Paris', 'Berlin', 'Amsterdam', 'London']})

for index, row in countries.iterrows():
  row['capital'].upper()
````
